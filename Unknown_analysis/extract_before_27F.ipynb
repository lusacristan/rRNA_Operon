```python
import pandas as pd
import re
```

```python
def extract_sequence_before_primer(fastq_file, primer_sequence, output_file):
    primer_pattern = primer_sequence.replace('M', '[AC]')
    
    with open(fastq_file, 'r') as input_file, open(output_file, 'w') as output_file:
        for line in input_file:
            if line.startswith('@'):  # Identifies the start of a sequence read
                sequence = next(input_file).strip()
                next(input_file)  # Skip the '+' line
                quality = next(input_file).strip()
                
                match = re.search(primer_pattern, sequence)
                if match:
                    extracted_sequence = sequence[:match.start()]
                    output_file.write(extracted_sequence + '\n')
                    print(extracted_sequence)

# Usage example
fastq_file = '/home/bcem2022/Documents/output_dir/rescued_unk.fq'
output_file = '/home/bcem2022/Documents/output_dir/output.txt'
primer_sequence = 'AGAGTTTGATCMTGGCTCAG'

extract_sequence_before_primer(fastq_file, primer_sequence, output_file)
```
